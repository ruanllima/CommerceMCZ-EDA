```python
import folium
import pandas as pd

# leitura dos dados da tabela
data = pd.read_csv("/content/maragogi.csv")
# criação do mapa
mapa = folium.Map(location=[data['place_lat'].mean(), data['place_lng'].mean()], zoom_start=65)
# adição dos marcadores de localização das empresas
for index, row in data.iterrows():
    folium.Marker(location=[row['place_lat'], row['place_lng']], tooltip=row['place_name']).add_to(mapa)

# exibição do mapa
mapa
```
